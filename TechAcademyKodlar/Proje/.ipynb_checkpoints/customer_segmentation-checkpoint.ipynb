##### Data Understanding

```python
import pandas as pd
df = pd.read_csv("Mall_Customers.csv")

df.head()
```

```python
# probleme baktığımızda elimizde bir hedef özellik yok, elimizdeki özellikleri kullanarak bizden bir gruplandırma yapmamız bekleniy
# yani bir unsupervised(label yok) tekniği kullanacağız
```

##### Data Preparation

```python
# öncelikli olarak elimizdeki verinin nan değeri, gereksiz sütunu var mı onun kontrolünü sağlayalım
df.info()
```

```python
df.columns
```

```python
# burada baktığımızda nan değer bulunmamaktadır
# özelliklere baktığımızda ise CustomerID bizim için herhangibir anlam ifade etmiyor, o sütunu çıkartarak daha doğru öğrenme sağlayabiliriz
df = df.drop(columns="CustomerID",axis = 1)
df.head()
```

```python
# Sütundaki tüm farklı değerleri listeler
print(df['Gender'].unique())
```

```python
# şimdi de kullanacağımız algortimaların daha rahat anlayabilmesi için öncelikli olarak gender kısmını sayısal değere çevirelim
# burada sadece Male-Female olduğu için Label Encoding, One-Hot-Encodinge göre daha avantajlı olacaktır
df["Gender"] = df["Gender"].map({"Male":1, "Female":0})
df.head()
```

##### Modeling

```python
# şimdi biz bir clustering(kümeleme) algoritması kullanacağız ve kullanabileceğimiz opsiyonlar
# K-means -> müşteri segmentasyonu, genel gruplama, resim sıkıştırma vs
# hierarchical -> hiyerarşinin önemli olduğu, biyoloji-tür sınıflandırmalarında vs
# DBSCAN -> coğrafi veriler, anamoli tespiti
# yani bizim için en uygun model K-means
```

```python
# burada öncelikli olarak verideki max-min vs değerlerine bakarak ölçeklendirme ihtiyacımız olup olmamasına bakalım
df.describe()
```

```python
# göreceğimiz üzere farklı özelliklerdeki değerler arasındaki farklar çok fazla ve bu modelin yanlış yorumlar-çıkarımlar yapmasına sebebiyet verebileceği
# için ölçeklendirme ile verileri aynı standart sayılar arasına sıkıştıracağız
from sklearn.preprocessing import StandardScaler
X = df.values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled[:10])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Korelasyon Matrisini Hesaplayalım
# Sadece analizde kullandığımız sayısal sütunları seçiyoruz
# Gender sütunu zaten 0 ve 1 olarak kodlanmıştı, o yüzden dahil edebiliriz.
features = ['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']
corr_matrix = df[features].corr()

# 2. Isı Haritası (Heatmap) Çizelim
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Özellikler Arası Korelasyon Matrisi (İlişki Haritası)', fontsize=14)
plt.show()
```

```python
# şimdi de K-means algortimasının hiperparametre ayarlamalarına bakalım
# n_cluster -> veriyi kaç parçaya böleceğimiz -> burada Elbow metod ile optimal değerleri bulacağız
# init -> başlangıçta hangi noktalar merkez olacak -> random ya da k-means++ vardır
# n_init -> algoritmanın farklı başlangıç noktaları ile kaç kez yeniden çalıştırılacağını verir
# max_iter -> merkez noktaları bumak için maksimum iterasyon sayısı
# random_state -> sonuçların her çalıştırmada aynı kalması için
```

```python
# burada öncelikli olarak n_cluster bulalım
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = [] # Hata kareler toplamını (Inertia) tutacağımız liste

# 1'den 10'a kadar küme sayılarını deniyoruz
for k in range(1, 11):
    # init='k-means++' : genelde bu tercih edilir
    # random_state=42  : Her seferinde aynı sonucu almak için
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_scaled) # Ölçeklenmiş veriyi kullanıyoruz!
    wcss.append(kmeans.inertia_) # Modelin hata değerini listeye ekliyoruz

# sonuçları görselleştirme
plt.figure(figsize=(10,5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method (Dirsek Yöntemi)')
plt.xlabel('Küme Sayısı (K)')
plt.ylabel('WCSS (Hata Skoru)')
plt.grid(True) # Grafiği okumayı kolaylaştırmak için ızgara ekledim
plt.show()
```

```python
# burada baktığımızda en net kırılım 2de ama oradaki hata oranı çok yüksek
# sonrasında ise 4 ile 6 da kırılımlar göze çarpıyor burada 4-5-6 hepsini deneyerek bir yol izleyelim
# alt taraftaki kodda her bir k değerinin skorunu karşılaştırıp en iyi olanı seçeceğiz

from sklearn.metrics import silhouette_score # -1 ile 1 arasında değer alıp, 1 e yakın olması modelin daha iyi olduğu anlamına gelir
import matplotlib.pyplot as plt
import seaborn as sns

# Denenecek K değerleri
k_values = [4, 5, 6]

# Grafikleri yan yana çizdirmek için alan oluşturuyoruz (1 satır, 3 sütun)
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

for i, k in enumerate(k_values):
    # 1. Modeli Kur ve Eğit
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    y_pred = kmeans.fit_predict(X_scaled) 
    
    # 2. Başarı Skorunu Hesapla (Silhouette Score)
    score = silhouette_score(X_scaled, y_pred)

    # 3. Görselleştirme (Orijinal Gelir ve Puan verisini kullanarak çiziyoruz ki anlaşılır olsun)
    # X ekseni: Yıllık Gelir (df'in 2. indeksi), Y ekseni: Harcama Skoru (df'in 3. indeksi)
    # df.iloc[:, 2] -> Annual Income (Gelir)
    # df.iloc[:, 3] -> Spending Score (Puan)
    sns.scatterplot(x=df.iloc[:, 2], y=df.iloc[:, 3], hue=y_pred, palette='viridis', ax=axes[i], s=100,legend = "full")
    # iloc ile bir karışıklık olmaması adına isim yazmak yerine kullanılır, kodun güvenle çalışması içindir 
    # hue -> renklendirmenin hangi değerlere göre olacağını verir
    # hue ile uygulanacak olan grupların renk paleti viridis
    # s yucarlakların büüklüğünü belirler
    # ax ile verinin nerede konumlandırılacağı belirlenir
    # legend ile full dediğimiz için grafiğin yanındaki grup numaralı tam gözükür
    
    # Merkezleri de çizelim (Ölçekli oldukları için ters dönüşüm yapmak zor, şimdilik sadece kümeleri görelim)
    axes[i].set_title(f'K = {k} | Siluet Skoru: {score:.3f}', fontsize=14, fontweight = "bold")
    axes[i].set_xlabel('Yıllık Gelir (k$)')
    axes[i].set_ylabel('Harcama Skoru (1-100)')
    axes[i].legend(title='Grup No', loc='best') # loc=best ile kutucuğun yerini python kendi belirler
    axes[i].grid(True, alpha=0.3) 

plt.tight_layout() # grafikler arası boşlukları kendisi otomatik ayarlar
plt.show()
```

```python
# burada görüleceği üzere değerler birbirine çok yakın olduğu için modellerden birini seçmemiz işimize yarayacaktır,
# burada hangisini seçeceğimiz ise, bizim verimiz için görüleceği üzere yıllık gelir ve harcama skoru kritik bir değer
# az kazanan az harcayan, az kazanan çok harcayan, çok kazanan az harcayan, çok kazanan çok harcayan ve orta kazanıp orta harcayan
# burada temelde dikkat ettiğimizde 
# 1. Modeli K=5 ile kuralım
kmeans5 = KMeans(n_clusters=5, init='k-means++', n_init=10, random_state=42)
y_pred5 = kmeans5.fit_predict(X_scaled)

# 2. Sonuçları Tabloya Dökelim
df['Cluster6_ID'] = y_pred6

# 3. Ortalamalara Bakalım (Özellikle AGE ve GENDER sütununa dikkat et)
summary6 = df.groupby('Cluster6_ID')[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
summary6['Kişi Sayısı'] = df['Cluster6_ID'].value_counts()

print(summary6)
```
