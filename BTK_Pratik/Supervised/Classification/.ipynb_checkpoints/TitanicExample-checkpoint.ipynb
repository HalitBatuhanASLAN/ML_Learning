Bu bÃ¶lÃ¼mde, verideki kategorik sÃ¼tunlarÄ± (Cinsiyet, Hayatta Kalma, Yolcu SÄ±nÄ±fÄ± vb.) inceliyoruz. AmaÃ§, her bir kategorinin veri seti iÃ§inde ne kadar yer kapladÄ±ÄŸÄ±nÄ± (frekansÄ±nÄ±) gÃ¶rmek ve bu deÄŸiÅŸkenlerin daÄŸÄ±lÄ±mÄ±nÄ± anlamaktÄ±r.

```python
def bar_plot(variable):
    var = train_df[variable]
    # benzersiz verilerin sayÄ±sÄ±nÄ± alÄ±yoruz
    varValues = var.value_counts()

    # geniÅŸlik 9, yÃ¼kseklik 3 olacak ÅŸekilde grafik boyutu belirlenmesi
    plt.figure(figsize=(9,3))
    # x ekseni kategoriler ve y ekseni deÄŸerler olacak ÅŸekilde grafiÄŸi oluÅŸtur
    plt.bar(varValues.index,varValues)
    # x eksenindeki yazÄ±larÄ±n dÃ¼zgÃ¼n gÃ¶rÃ¼nmesi iÃ§in ayarlamalar
    plt.xticks(varValues.index,varValues.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValues))
```

## Outlier Detection

```python
def detect_outliers(df,features):
    outlier_indices = []

    for c in features:
        #first quartile( 1. Ã§eyrek %25'lik dilim)
        Q1 = np.percentile(df[c],25)
        #third quartile( 3. Ã§eyrek %75'lik dilim)
        Q3 = np.percentile(df[c],75)
        # IQR Ã§eyrekler aÄŸÄ±rlÄ±ÄŸÄ±
        IQR = Q3 - Q1
        # outlier step genelde IQR'Ä±n 1.5 katÄ± olur
        outlier_step = IQR * 1.5
        #detect outliers and their indeces -> alt ve Ã¼st sÄ±nÄ±r dÄ±ÅŸÄ±ndakiler bulunur
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        #store indeces
        outlier_indices.extend(outlier_list_col)
    # hangi satÄ±rÄ±n kaÃ§ kez aykÄ±rÄ± deÄŸer olarak iÅŸaretlendiÄŸi sayÄ±lÄ±r
    outlier_indices = Counter(outlier_indices)
    # birden Ã§ok sÃ¼tunda aykÄ±rÄ± deÄŸerler tespit edilir 
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 2)

    return multiple_outliers
```

```python
train_df.loc[detect_outliers(train_df,["Age","SibSp","Parch","Fare"])]
# bunlar arasÄ±nda outlier satÄ±rlarÄ± olanlarÄ±n tespiti 
```

```python
train_df = train_df.drop(detect_outliers(train_df,["Age","SibSp","Parch","Fare"]),axis = 0).reset_index(drop = True)
# ilgili satÄ±rlarÄ± Ã§Ä±kart
```

```python
random_state = 42
classifier = [DecisionTreeClassifier(random_state = random_state),
             SVC(random_state = random_state),
             RandomForestClassifier(random_state = random_state),
             LogisticRegression(random_state = random_state),
             KNeighborsClassifier()]

#  bir dÃ¼ÄŸÃ¼mÃ¼n bÃ¶lÃ¼nmesi iÃ§in gerekli min Ã¶rnek sayÄ±sÄ± 10 dan 500 kadar 20ÅŸer artÄ±rarak dener
dt_param_grid = {"min_samples_split" : range(10,500,20),
                "max_depth":range(1,20,2)}

# kernel veri ayrÄ±mÄ±ndaki kullanÄ±lacak matsal fonksiyon tipini rbf olarak belirledik
# tek bir eÄŸitim Ã¶rneÄŸinin ne kadar uzaa etki edeceÄŸini belirler
# bÃ¼yÃ¼k C daha az hataya odaklanÄ±r, burada ne kadar tolerans edileceÄŸi verilmiÅŸ
svc_param_grid = {"kernel": ["rbf"],
                 "gamma": [0.001,0.01,0.1,1],
                 "C": [1,10,50,100,200,300,1000]}

# max_features aÄŸaÃ§ bÃ¶lÃ¼nÃ¼rken bakÄ±lacak maksimum Ã¶zellik sayÄ±sÄ±
# dallanma ve yaprak sayÄ±sÄ± iÃ§in gereken Ã¶rnek sayÄ±larÄ±
# verilerin tamamÄ± seÃ§ilir false olduÄŸu iÃ§in, yoksa rastgele seÃ§ilirdi
# bÃ¶lÃ¼nmenin kalitesini Ã¶lÃ§en matematiksel Ã¶lÃ§Ã¼t gini safsÄ±zlÄ±ÄŸÄ±
rf_param_grid = {"max_features":[1,3,10],
                "min_samples_split":[2,3,10],
                 "min_samples_leaf":[1,3,10],
                 "bootstrap":[False],
                 "criterion":["gini"]}

# C regÃ¼lariszyonun(aÅŸÄ±rÄ± Ã¶ÄŸrenmeyi engelleme) tersidir, 10^-3 ile 10^3 arasÄ± 7 farklÄ± deÄŸer alÄ±r
# modelin karmaÅŸÄ±klÄ±ÄŸÄ±nÄ± L1 veya L2 ile cezalandÄ±rma
logreg_param_grid = {"C":np.logspace(-3,3,7),
                    "penalty":["l1","l2"]}

# bakÄ±lacak en yakÄ±n komÅŸu sayÄ±sÄ± 1-19 arasÄ±
# yakÄ±n komÅŸularÄ±n oy gÃ¼cÃ¼nÃ¼n uzaklÄ±ÄŸa gÃ¶re deÄŸiÅŸip deÄŸiÅŸmeyeceÄŸi
# iki nokta arasÄ± mesafeyi Ã¶lÃ§me yÃ¶ntemi
knn_param_grid = {"n_neighbors":np.linspace(1,19,10,dtype = int).tolist(),
                 "weights":["uniform","distance"],
                 "metric":["euclidean","manhattan"]}

classifier_param = [dt_param_grid,
                   svc_param_grid,
                   rf_param_grid,
                   logreg_param_grid,
                   knn_param_grid]
```

```python
cv_result = [] # her bir modelin en iyi baÅŸarÄ± skorunu tutmak iÃ§in
best_estimators = [] # her modelin en iyi optimize edilmiÅŸ parametrelerini tutmak iÃ§in
for i in range(len(classifier)):
    # GridSearch ile her bir classifier iÃ§in iÃ§erdiÄŸi tÃ¼m kombinasyonlarÄ± deniyoruz, n_splits ile veriyi 10 bÃ¶lÃ¼p 9 ile eÄŸitim, 1 ile test(coss validation)
    # yapÄ±p, baÅŸarÄ±nÄ± accuracy deÄŸerine gÃ¶re ayarlayÄ±p, n_jobs-1 ile bilgisayardaki iÅŸlemci Ã§ekirdeklerinden kullanÄ±labileceklerin hepsini kullanÄ±r bÃ¶ylece daha hÄ±zlÄ± olur
    # verbose ile de iÅŸlem sÄ±rasÄ±ndaki denene kombinasyonlar ne kadar kaldÄ±ÄŸÄ± igbi anlÄ±k bilgileri yazar
    clf = GridSearchCV(classifier[i],param_grid = classifier_param[i], cv = StratifiedKFold(n_splits = 10),scoring = "accuracy",n_jobs = -1,verbose = 1)
    clf.fit(X_train,y_train)
    cv_result.append(clf.best_score_)
    best_estimators.append(clf.best_estimator_)
    print(cv_result[i])
```

```python
cv_results = pd.DataFrame({"Cross Validation Accuracy Means":cv_result, "ML models":["DecisionTreeClassifier","SVM","RandomForest",
                        "LogisticRegression",
                        "KNeighborsClassifier"]})
cv_results
```
