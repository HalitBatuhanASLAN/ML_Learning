```python
d = pd.DataFrame(df["Potability"].value_counts()) #her değerin kaçar edet olduğunu 
# hole ile ortada bir boşluk bırakılır
# opacity çemberlerin opaklığını ayarlar
fig = px.pie(d, values = "count", names= ["Not Potable","Potable"], hole = 0.35, opacity = 0.8,
            labels = {"label": "count","Potability":"Number of Samples"})
fig.update_layout(title = dict(text = "Pie chart of potability feature"))
fig.update_traces(textposition = "outside", textinfo = "percent+label")
fig.show()
```
